## pykrx

- 1990년 데이터부터 조회 가능

```python
#10개년치 주가 정보 수집

start_date = '2012-05-17'
end_date = '2022-05-17'

df_pykrx = stock.get_market_ohlcv_by_date(fromdate=start_date, 
                                          todate=end_date,
                                          ticker="005930")
df_pykrx
```

```python
#10개년치 주가 정보 수집 adjusted=False 를 추가하면 거래대금과 등락률이 추가됨

start_date = '2012-05-17'
end_date = '2022-05-17'

df_pykrx = stock.get_market_ohlcv_by_date(fromdate=start_date, 
                                          todate=end_date,
                                          ticker="005930", adjusted=False)
df_pykrx
```

## yfinance
- yahoo finance를 데이터 소스로 사용하는 패키지이기 때문에 위에서 설명한 바와 같이 코스피 종목에는 .KS, 코스닥 종목에는 .KQ를 붙여줘야 함
- 가장 다양한 정보 확인 가능 - 배당정보, 분할정보, 재무정보 등
- end에 설정한 일자의 전일자까지 조회되기 때문에 조회하고자 하는 "종료일+1일을 end에 넣어줘야함"
- 분단위 데이터도 조회 가능하지만 조회 가능한 기간에 제한이 있음

```python
start_date = '2012-05-17'
end_date = '2022-05-18' #17일 까지의 데이터를 수집하고 싶어 end_date를 18일로 수정한다.

ticker = yf.Ticker('005930.KS')

ticker.history(
               interval='1d',
               start=start_date,
               end=end_date,
               actions=True,
               auto_adjust=True)
```

```python
import pandas as pd
import numpy as np
import time, json, requests 
from datetime import datetime, timedelta, date

import pymysql # python에서 mysql을 사용하는 패키지

import sqlalchemy # sql 접근 및 관리를 도와주는 패키지
from sqlalchemy import create_engine
```

```python
server = '127.0.0.1' # local server
user = 'root' # user name
password = '1541' # 개인 password
db = 'myfirst_db' # DB 이름

# sqlalchemy의 create_engine을 이용해 DB 연결
engine = create_engine('mysql+pymysql://{}:{}@{}/{}?charset=utf8'.format(user,password,server,db))
```

```python
# FinanceDataReader를 이용해 수집한 데이터를 DB에 저장

df_fdr.to_sql(name='kospi_naver',con=engine,if_exists='append',index=False,
             dtype = { # sql에 저장할 때, 데이터 유형도 설정할 수 있다.
                 
                 'Open' : sqlalchemy.types.BIGINT(),
                 'High' : sqlalchemy.types.BIGINT(),
                 'Low' : sqlalchemy.types.BIGINT(),
                 'Close' : sqlalchemy.types.BIGINT(),
                 'Volume' : sqlalchemy.types.BIGINT(),
                 'Change' : sqlalchemy.types.FLOAT(),
                 'Date' : sqlalchemy.types.DATE(),
             }
            )
```
